```python
#Importing libraries
import pandas as pd
import numpy as np
import requests
import urllib
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup
from datetime import datetime
import os

# Passing on user inputs to defined function for url extraction
def get_results(search_terms, no_of_results, fname, save_file=True):
    query = search_terms 
    query = urllib.parse.quote_plus(query) 
    result_no = no_of_results
    links_list = []
    ctr = 0
    now = datetime.now()
    time_string = str(now.strftime("%Y-%m-%d %H:%M:%S"))
     
    '''Setting a user agent i.e. A ‘User-Agent’ HTTP request header is a string that a web browser
    is sending to a web server along with each request to identify itself'''
    headers = {'User-Agent': 'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:77.0) Gecko/20100101 Firefox/77.0'}
    
    #Making the GET request and creating the Soup
    url = "https://www.google.com/search?q=" + query + "&num=" + str(result_no)
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")
    search_results = soup.select('.g a')
    
    #Generating output text files for lists of pdf & html urls
    pdf = open("pdf.txt","w")
    htm = open("htm.txt","w")
       
    #Consolidating url lists based on type (html vs. pdf)
    for link in search_results[:]:
        if ctr < int(result_no):
            url_link = link.get('href')
            if url_link is None:
                break
            elif url_link.startswith('#') or \
                url_link.startswith('https://webcache.googleusercontent.com') or \
                url_link.startswith('/search?'):   
                continue
            elif url_link.__contains__("wikipedia") == True:
                continue
            elif url_link.__contains__("support.google") == True:
                continue
            else:
                links_list.append(url_link)
                ctr += 1
                if url_link.lower().endswith(".pdf"):
                    pdf.write(url_link + "\n")
                else:
                    htm.write(url_link + "\n")
        else:
            break
 
    # Conversion of urls list to data frame
    df_links = pd.DataFrame(links_list)
            
    # CSV output file saved to the current folder   
    filename = fname + "_" + str(datetime.now().strftime("%Y%m%d-%H%M%S"))+" "+ query    
    df_links.to_csv(os.path.join(r"..", filename), index=False)
    
    return links_list
            
# User input for search terms and number of results to be returned        
search_terms = str(input("Enter search terms separated by a space: ")[:200])
no_of_results = int(input("Enter number of results to be returned: "))
fname = (input('Enter filename: '))
get_results(search_terms, no_of_results,fname)     
```
